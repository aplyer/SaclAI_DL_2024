### Un générateur

Simplement une fonction retournant une valeure à l'appelant en intérompant ses itérations

```python
def myrange_square(val):
    for i in range(val):
        yield i*i

for i in myrange_square(4):
    print(i)
```

Un générateur peut aussi être utilisé *manuellement*

```python
gen = myrange_square(42)
print(next(gen), next(gen), next(gen), next(gen))
```

### Les classes

```python
class Voiture:
    def __init__(self, marque = ''):
        self.marque = marque
        self._compteur = 0
    def ajouteKm(self, km):
        self._compteur += km
    def retourneCompteur(self):
        return self._compteur

maVoiture = Voiture('Wolkwagen')
maVoiture.ajouteKm(200000)
```

Tout les membre d'une classe sont accessible, cependant une convention existe :
pour signifier un membre privé ou protégé on le préfix de deux ou un underscore : _

```python
print(maVoiture._compteur)
#si les accesseur existe mieux vaut les utiliser
print(maVoiture.retourneCompteur())
```

### toute la souplesse de python

Les lambda expression permettent de traffiquer des instance de classes en leurs rajoutant dynamiquemet des méthodes!

```python
def traffique(voiture,km):
    voiture._compteur -= km

maVoiture.traffique = lambda x : traffique(maVoiture, x)

maVoiture.traffique(100000)

print(maVoiture.retourneCompteur())
```

plutot utile si je veux pouvoir vendre ma voiture!

## Les exceptions

Avec toute cette souplesse il est important de prévoir une gestion efficace des érreures à l'exécution : les exceptions sont la pour ca!

```python
def foo(x):
    if type(x) is type(1):
        print(x)
    else:
        raise TypeError('x doit etre un entier')
```

```python
try:
    foo(1)
except TypeError:
    print(e)
else:
    print('tout c\'est bien passé!')
```

```python
try:
    foo('a')
except TypeError as e:
    print(e)
else:
    print('tout c\'est bien passé!')
```

```python
def foo(x):
    if type(x) is type(1):
        if x > 0:
            print(x)
        else:
            raise ValueError('x doit etre positif')
    else:
        raise TypeError('x doit etre un entier')
```

```python
try:
    foo(-1)
except TypeError as e:
    print(e)
else:
    print('tout c\'est bien passé!')
```

oups...

```python
try:
    foo(-1)
except TypeError as  e:
    print('exception de type : ', e)
except ValueError as e:
    print('exception de valeure : ',e)
else:
    print('tout c\'est bien passé!')
```

les exception doivent être récupérer... si une exception remonte au niveau de l'interpreteur on a un beau message d'erreur.
